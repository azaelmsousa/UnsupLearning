# Reading the data

The data is a series of news posted on Twitter regarding health. To process this data, it must be transformed into a feature vector. To do that, two methods were applied: Bag of Words and Word2Vec.

The Bag of Words (BOW) descriptor creates a long vector (ideally with the size of all possible words in the english lexicon) where each word corresponds to a position of this vector. If a word is repeated in a sentence, then the value in the position of the vector corresponding to that word is increased by 1.

The Word2Vec descriptor tries to define a certain relationship between words based on its neighbors in the training data. For instance, the words "king" and "man" appear more often in the same sentence than any of them with the words "trip" and "bag". Thus, the Word2Vec define similar vectors for words with similar meaning.

# Deciding which descriptor to use

This is the first experiment of this project and aims to define the number of groups (or clusters) formed by the data. In order to have a first glimpse of how the data is scattered in the high-dimentional space, we applied the t-Distributed Stochastic Neighbor Embedding (t-SNE) algorithm.

```python
%matplotlib inline
import os
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np
from glob import glob
import matplotlib.cm as cm
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, RobustScaler

smp_sz = 7000
```

```python
def visualize_scatter(data_2d, figsize=(20,20)):
    plt.figure(figsize=figsize)
    plt.grid()
    
    plt.scatter(data_2d[:,0],data_2d[:,1],marker='o',linewidth='1',alpha=0.8)    
```

## Projecting the BOW data
Here, the BOW data is projected using the t-SNE. Note that the data form a huge cluster in the middle, with a few other smaller clusters around it. Also, some datapoints form a sort of crescent moon.

```python
from MulticoreTSNE import MulticoreTSNE as TSNE

tsne_bow = TSNE(n_components=2, perplexity=500, verbose=True, n_jobs=-1)
tsne_bow_result = tsne_bow.fit_transform(bow[:smp_sz])
```

```python
visualize_scatter(tsne_bow_result,(30,30))
```

## Projecting the Word2Vec data
Here, the Word2Vec data is projected using the t-SNE. Note that most of the data form a single cluster. Despite some clusters can be seen, it would be difficult to separate them precisely.

```python
tsne_word2vec = TSNE(n_components=2, perplexity=500, verbose=True, n_jobs=-1)
tsne_word2vec_result = tsne_word2vec.fit_transform(word2vec)
```

```python
visualize_scatter(tsne_word2vec_result,(30,30))
```

## Chosen Descriptor
After looking on both projections, the Bag of Words descriptor seems to be more sensitive to the kind of data explored in this work. However, more experiments must be performed to assess the descriptors.
